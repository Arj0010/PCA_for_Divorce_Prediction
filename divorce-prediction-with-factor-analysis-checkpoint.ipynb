### Factor Analysis

Factor analysis is a linear statistical model, that is used to explain the variance among the observed variable and condense a set of the observed variable into the unobserved variable called factors. Observed variables are modeled as a linear combination of factors and error terms. Factor or latent variable is associated with multiple observed variables, who have common patterns of responses. Each factor explains a particular amount of variance in the observed variables. It helps in data interpretations by reducing the number of variables. ([Source](https://www.datacamp.com/community/tutorials/introduction-factor-analysis))

### Why Factor Analysis and not PCA?

1. Both Principal Components Analysis (PCA) and Factor Analysis are dimension reduction techniques.
2. Principal Component analysis makes the components that are completely orthogonal to each other whereas Factor analysis does not require such the factors to be orthogonal
3. Principal Components is primarily an exploratory technique to rationalize the number of variables in an analysis (for further causal analysis). Factor analysis, on the other hand, is approached with a prior knowledge or belief that which variables may be similar and therefore should load on a single factor. Factor analysis is thus a popular variable reduction technique in market research.


Principal Component Analysis (PCA) and Factor Analysis (FA) are both techniques used in multivariate analysis to reduce the dimensionality of data and identify underlying patterns or structures. The distinction you mentioned refers to how these methods treat the relationships between the extracted components or factors:

Principal Component Analysis (PCA):

PCA aims to create new variables, called principal components, that are linear combinations of the original variables. These components are constructed to be orthogonal to each other, meaning they are independent and uncorrelated.
The principal components are ordered by the amount of variance they explain in the data. The first component captures the most variance, the second component captures the second most variance, and so on.
PCA does not take into account the original variables or their meanings when creating the components. It only focuses on maximizing variance and creating orthogonal components.
Factor Analysis (FA):

Factor Analysis also aims to create new variables, called factors, that explain the underlying structure in the data. However, unlike PCA, FA allows the factors to be correlated with each other.
Factors in FA are not necessarily orthogonal. This means that they can have correlations with each other, indicating that they are related or represent similar underlying constructs.
FA assumes that observed variables are influenced by one or more latent (unobserved) factors. These factors can be correlated because they may capture overlapping aspects of the underlying structure of the data.
In summary, while PCA creates orthogonal components that are independent of each other, Factor Analysis allows factors to be correlated, reflecting the potential interrelationships or dependencies among the underlying constructs in the data. This difference in approach makes PCA more suitable for variance maximization and dimensionality reduction, while FA is often used for exploring underlying structures and relationships among variables.

#### Bartlett's test of sphericity 
-Bartlett’s test checks whether the correlation is present in the given data. It tests the null hypothesis (H0) that the correlation matrix is an Identical matrix. The identical matrix consists of all the diagonal elements as 1. So, the null hypothesis assumes that no correlation is present among the variables.

H0 : Correlation matrix is an Identical matrix

H1 : Correlation matrix is not Identical matrix

```python
data = df.drop('Divorce_Y_N',axis=1)
```

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi2,p = calculate_bartlett_sphericity(data)
#built in function to calculate bartlett test
print("Chi squared value : ",chi2)
print("p value : ",p)
```

**The p-value < 0.05, correlation is present among the variables with 95% confidence interval**

### KMO Test

KMO Test measures the proportion of variance that might be a common variance among the variables. Larger proportions are expected as it represents more correlation is present among the variables thereby giving way for the application of dimensionality reduction techniques such as Factor Analysis. **KMO score is always between 0 to 1 and values more than 0.6 are much appreciated.**

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_vars,kmo_model = calculate_kmo(data)
print(kmo_model)
```

**KMO score is close to 1, thus applying factor analysis can be effective for the dataset.**

**SCREE PLOT**

The scree plot is a graphical tool that helps in determining the number of factors to retain in factor analysis based on the eigenvalues.It allows you to visually inspect the eigenvalues to determine the number of factors to retain in your factor analysis. Typically, factors corresponding to eigenvalues greater than 1 are considered for retention (refers to the process of deciding how many factors to keep or retain from the analysis.).

It displays the eigenvalues for each factor in descending order. The point at which the plot levels off (forming an "elbow") indicates the number of factors to retain. Factors before the elbow are retained, while those after the elbow are discarded.

```python
from factor_analyzer import FactorAnalyzer
n = data.shape[1]
fa = FactorAnalyzer(rotation = None,impute = "drop",n_factors=n)
fa.fit(data)
ev,_ = fa.get_eigenvalues()
plt.scatter(range(1,n+1),ev)

plt.plot(range(1,n+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigen Value')
plt.grid()
```

### Factor Loading

Factor loading can tell us how each factor created is explaining each of the variable. More value says that factor explains well about that variable. Sorting the factor load values helps us understand which variables are grouped to represent a new factor.

```python
fa = FactorAnalyzer(n_factors=5,rotation='varimax')
#varimax maximizes the variance of the squared loadings within each factor, making the factors more interpretable.
c=fa.fit(data)
fa_load = pd.DataFrame(fa.loadings_,index=data.columns)
fa_load
#so here the columns became the 5 new factors 0-4 after factor loading and the rows are the reasonn
```

```python
def highlight_max(s):
    # Get 5 largest values of the column
    is_large = s.nlargest(10).values
    # Apply style is the current value is among the 5 biggest values
    return ['background-color: black' if v in is_large else '' for v in s]

fa_load.sort_values(by=0,ascending=False).style.apply(highlight_max)
```
